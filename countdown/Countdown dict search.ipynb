# [8 of 10 Cats play] Count Down Word search.

Fragile search of *unix system dict word list and greps down possible words.

```python
%%bash
# Given letters: mgnmeoeae
# =>
# unique: mgneoa
# triple repeat: e
# double repeat: m
# single: gnoa

cat /usr/share/dict/words | grep -E '^[mgnmeoeae]*$' | grep -v -E '([e]).*\1{3}' | grep -v -E '([m]).*\1{2}' | grep -v -E '([aong]).*\1' | awk '{ print length($0) " " $0; }' $file | sort -r -n 
```
